````python
sponsored_links = ''
non_sponsored_links = ''

link = 'https://www.ebay.com/sch/i.html?_nkw=playstation+4+slim&LH_BIN=1&_ipg=100'
# find the sponsored items using Regex and print links
for pg_num in tqdm(range(1,11)):
    new_link = '{}&_pgn={}'.format(link, pg_num)
    print(new_link)
    res = requests.get(new_link)
    soup = bs.BeautifulSoup(res.content, 'html.parser')
    all_divs = soup.find_all('a',class_='s-item__link')
    
    for div_block in all_divs:
        if re.search(r'.S.*P.*O.*N.*S.*O.*R.*E.*D.', str(div_block)) is None:
            non_sponsored_link = div_block.get('href')    
            if not non_sponsored_links:
                non_sponsored_links = non_sponsored_link
            non_sponsored_links = non_sponsored_links + '\n' + non_sponsored_link
        else: 
            sponsored_link = div_block.get('href')
            if not sponsored_links:
                sponsored_links = sponsored_link
            sponsored_links = sponsored_links + '\n' + sponsored_link
````

````python
with open('sponsored.txt','wb') as open_file:
        open_file.write(sponsored_links.encode('utf-8'))
        open_file.close()
with open('non-sponsored.txt','wb') as open_file:
        open_file.write(non_sponsored_links.encode('utf-8'))
        open_file.close()
````

````python
def extract_title(soup):
    """
    Extract Product Title

    This function extracts product title taken from a parsed html file called soup.
    It iterates through two search options – one for regular pages and one for specially featured pages
    """
    title = None
    for code in (
        # extracting title for regular pages
        lambda: soup.find(id='itemTitle').contents[-1],
        # extract title for hero product / featured product
        lambda: soup.find('h1', class_='product-title').get_text(),
    ):
        try:
            title = code()
        except:
            continue
        break 
    else:
        print('Warning: No Title Extracted, set to None')
        
    return title
````

````python
def has_best_offer(soup):
    """
    Extract Best Offer

    This function determines if product has best offer taken from a parsed html file called soup.
    """
    best_offer = False
    try:
        if re.search('Best Offer:', soup.find(id='mainContent').get_text()) is not None:
            best_offer = True
    except:
        print('Warning: No Best Offer Extracted, set to False')
        
    return best_offer
````

````python
def returns_arrangement(soup):
    """
    Extracts Arrangement for Returns

    This function determines if a seller allows for item returns taken from a parsed html file called soup.
    """
    returns = None
    try:
        returns = soup.find(id='vi-ret-accrd-txt').get_text()
    except:
        print('Warning: No Returns Extracted, set to None')
    
    if returns is not None:
        if 'NOT ACCEPT' in returns.upper():
            returns = False
        else:
            returns = True
        
    return returns
````

````python
def extract_shipping_price(soup):
    """
    Extracts Shhipping Price

    This function determines shipping price of an item from a parsed html file called soup.
    It iterates through four possible scenarios getting all US Dollar shipping price for a specific item.
    """
    shipping_price = None
    for code in (
        # catering for multiple currency shipping
        lambda: re.sub(r'[^\d.]', '', re.findall(
                    r'US \$[\d]+.[\d]*|\$[\d]+.[\d]*', soup.find(id='shSummary').findNext('span').get_text())[-1]),
        lambda: re.sub(r'[^\d.\w ]', '', soup.find(id='shSummary').contents[1].get_text()),
        lambda: re.sub(r'[^\d.]', '', soup.find(id='shSummary').findNext('span').get_text()),
        lambda: soup.find('span', class_='logistics-cost').get_text(),
    ):
        try:
            shipping_price = code()
        except:
            continue
        break 
    else:
        print('Warning: No Shipping Price Extracted, set to None')
    
    if 'FREE' in str(shipping_price).upper():
        # set price to zero
        shipping_price = 0
    elif any(stopword in str(shipping_price).upper() for stopword in ['CALCULATE', 'LOCAL']):
        # assign a number worth noticing to fix later
        shipping_price = -99999
        
    try:
        shipping_price = int(float(shipping_price) * 100)
    except:
        shipping_price
        
    return shipping_price
````

````python
def extract_item_condition(soup):
    """
    Extracts Item Condition

    This function determines item condition from a parsed html file called soup.
    """
    item_condition = None
    
    try:
        item_condition = soup.find(id='vi-itm-cond').get_text()
    except:
        print('Warning: No Item Condition Extracted, set to None')
    
    return item_condition
````

````python
def extract_seller_name(soup):
    """
    Extracts Seller Name

    This function determines seller name from a parsed html file called soup.
    """
    seller_name = None
    
    try:
        seller_name = soup.find_all('span', {'class': 'mbg-nw'})[0].get_text()
    except:
        print('Warning: No Seller Name Extracted, set to None')
    
    return seller_name
````

````python
def extract_seller_score(soup):
    """
    Extracts Seller Score

    This function determines seller score from a parsed html file called soup.
    """
    seller_score = None
    try:
        seller_score = re.sub(r'[^\d]', '', soup.find_all('span', {'class': 'mbg-l'})[0].get_text())
    except:
        print('Warning: No Seller Score Extracted, set to None')
    
    try:
        seller_score = int(seller_score)
    except:
        seller_score
        
    return seller_score
````

````python
def extract_items_sold(soup):
    """
    Extracts Items Sold

    This function determines items sold from a parsed html file called soup.
    """
    items_sold = None
    try:
        items_sold = re.search('[\d]+(?= sold)', soup.find(id='mainContent').get_text()).group()
    except:
        items_sold = None
    
    try:
        items_sold = int(items_sold)
    except:
        items_sold
    
    return items_sold
````

````python
def extract_item_price(soup):
    """
    Extracts Item Price

    This function determines item price from a parsed html file called soup.
    """
    item_price = None
    # this is the standard implementation of getting the price
    for code in (
        lambda: re.sub(r'[^\d.]', '', soup.find(id='prcIsum').get_text()),
        # this is the implementation of getting price for items under different currency
        lambda: re.sub(r'[^\d.]', '', soup.find(id='convbidPrice').get_text()),    
        # this is the implementation of getting price under discount
        lambda: re.sub(r'[^\d.]', '', soup.find(id='mm-saleDscPrc').get_text()),         
        # this is the implementation of getting price for hero items  
        lambda: re.sub(r'[^\d.]', '', soup.find('h2', class_='display-price').get_text())
    ):
        try:
            item_price = code()
        except:
            continue
        break
    else:
        print('Warning: No Items Price Extracted, set to None')
    
    try:
        item_price = int(float(item_price) * 100)
    except:
        item_price
        
    return item_price 
````

````python
def get_information(source):
    mypath = './{}'.format(source)
    files = [f for f in listdir(mypath) if isfile(join(mypath, f))]
    items = pd.DataFrame(files, columns=['item_id'])
    items['item_id'] = items['item_id'].str.replace('.htm', '')
    items['is_sponsored'] = source.upper() == 'SPONSORED'
    for file_name in tqdm(files):
        item_id = re.sub(r'[^\d]', '', file_name)
        with open('./{}/{}'.format(source, file_name), 'r') as open_file:
            # parse the page and find appropriate tag
            try:
                soup = bs.BeautifulSoup(open_file, 'html.parser')
            except:
                try: 
                    soup = bs.BeautifulSoup(open_file, from_encoding='iso-8859-8')
                except: 
                    print('item skipped')
                    continue
            
            items.loc[items['item_id'] == item_id, 'title'] = extract_title(soup)
            
            items.loc[items['item_id'] == item_id, 'best_offer'] = has_best_offer(soup)
            
            items.loc[items['item_id'] == item_id, 'returns_accepted'] = returns_arrangement(soup)
            
            items.loc[items['item_id'] == item_id, 'shipping_price'] = extract_shipping_price(soup)

            items.loc[items['item_id'] == item_id, 'item_condition'] = extract_item_condition(soup)

            items.loc[items['item_id'] == item_id, 'seller_name'] = extract_seller_name(soup)

            items.loc[items['item_id'] == item_id, 'seller_score'] = extract_seller_score(soup)

            items.loc[items['item_id'] == item_id, 'items_sold'] = extract_items_sold(soup)
            
            items.loc[items['item_id'] == item_id, 'price'] = extract_item_price(soup)

    items.loc[items['shipping_price'] < 0, 'has_variable_shipping_price'] = True
    items.loc[items['shipping_price'] >= 0, 'has_variable_shipping_price'] = False
    items.loc[items['shipping_price'] < 0, 'shipping_price'] = None
    return items
````

````python
non_sponsored_items = get_information('non-sponsored')
````

````python
sponsored_items = get_information('sponsored')
````

````python
all_items = pd.concat([sponsored_items, non_sponsored_items])
````

````python
%%sql

SELECT is_sponsored, 
item_condition, 
AVG(shipping_price / 100) AS AverageShippingPrice, 
MAX(shipping_price / 100) AS MaxShippingPrice,
MIN(shipping_price / 100) AS MinShippingPrice
FROM eBay_items
WHERE shipping_price IS NOT NULL
GROUP BY is_sponsored, item_condition;
````

````python
%%sql 

SELECT is_sponsored, 
item_condition, 
AVG(price / 100) AS AveragePrice, 
MAX(price / 100) AS MaxPrice,
MIN(price / 100) AS MinPrice
FROM eBay_items
WHERE price IS NOT NULL
GROUP BY is_sponsored, item_condition;
````

````python
%%sql

SELECT is_sponsored, 
item_condition, 
AVG(best_offer) AS AverageBestOffer, 
MAX(best_offer) AS MaxBestOffer,
MIN(best_offer) AS MinBestOffer
FROM eBay_items
WHERE best_offer IS NOT NULL
GROUP BY is_sponsored, item_condition;
````

````python
%%sql

SELECT is_sponsored, 
item_condition, 
AVG(returns_accepted) AS AverageReturns, 
MAX(returns_accepted) AS MaxReturns,
MIN(returns_accepted) AS MinReturns
FROM eBay_items
WHERE returns_accepted IS NOT NULL
GROUP BY is_sponsored, item_condition;
````

````python
%%sql 

SELECT is_sponsored, 
item_condition, 
AVG(seller_score) AS AverageSellerScore, 
MAX(seller_score) AS MaxSellerScore,
MIN(seller_score) AS MinSellerScore
FROM eBay_items
WHERE seller_score IS NOT NULL
GROUP BY is_sponsored, item_condition;
````
